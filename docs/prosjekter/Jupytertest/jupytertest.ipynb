# Eksempel på et program som som plotter en funksjon

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, 'ro-')
plt.axis([0, 6, 0, 30])
plt.show()
```

## Oppsummering av funksjoner

Som man kan se av formen på kurven så er dette en andregradsfunksjon. Denne kan vi skrive på formen $f(x)=x^2$ når i intervallet $x\in[1,5]$ i dette eksempelet.
